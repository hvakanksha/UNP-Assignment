```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

**Problem-1**<br>
In the “fmri” dataset, find  the range of the “timepoint” column using a box plot.

```python
x=sns.load_dataset('fmri')
```

```python
z = x["timepoint"]
sns.boxplot(z)
```

**Problem-2**<br>
Construct a pie chart describing the distribution of subclasses in the column “ region”.

```python
s = x.groupby('region')['region'].count()
```

```python
plt.pie(s, labels=s.index, autopct="%.2f%%");
plt.show()
```

**Problem-3**<br>
In the diamonds dataset, fit a regression plot for price Vs carat. (Use the appropriate order to fit the data properly)


```python
x=sns.load_dataset('diamonds')
```

```python
sns.regplot('price', 'carat', data=x.sample(1000), order=2, marker="4");
```

**Problem-5**<br>
In the penguins dataset, construct a violin plot of flipper length based on gender. Construct a boxen plot for the same, and determine if both the plots describe the data similarly.


```python
x=sns.load_dataset('penguins')
```

```python
sns.violinplot(x='flipper_length_mm', y='sex', data=x);
```

```python
sns.boxenplot(x='flipper_length_mm', y='sex', data=x);
```

**Problem-6**<br>
Construct a scatterplot for body mass Vs flipper length. Classify the data points based on the type of its species.(use hue)


```python
sns.scatterplot(x='body_mass_g', y='flipper_length_mm', hue='species', data=x);
```
